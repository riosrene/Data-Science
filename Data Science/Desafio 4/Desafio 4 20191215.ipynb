```python
#Cargamos librerias para empezar con KNN
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
X=df3
y=pd.Series(df3.FAVORABILIDAD)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=12)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=12)

scores_para_df = []

for i in range(3,5):
    
    # En cada iteración instanciamos el modelo con un hiperparámetro distinto
    model = KNeighborsClassifier(n_neighbors=i)
    
    # cross_val_scores nos devuelve un array de 5 resultados,
    # uno por cada partición que hizo automáticamente CV
    cv_scores = cross_val_score(model, X_train, y_train, cv=kf)
    
    # Para cada valor de n_neighbours, creo un diccionario con el valor
    # de n_neighbours y la media y el desvío de los scores.
    dict_row_score = {'score_medio':np.mean(cv_scores),\
                      'score_std':np.std(cv_scores), 'n_neighbours':i}
    
    # Guardo cada uno en la lista de diccionarios
    scores_para_df.append(dict_row_score)
```

```python
# Creamos el DataFrame a partir de la lista de diccionarios
df_scores = pd.DataFrame(scores_para_df)
df_scores.head()
```

```python
df_scores['limite_inferior'] = df_scores['score_medio'] - df_scores['score_std']
df_scores['limite_superior'] = df_scores['score_medio'] + df_scores['score_std']
df_scores.head()
```

```python
plt.plot(df_scores['n_neighbours'],df_scores['limite_inferior'],color='r')
plt.plot(df_scores['n_neighbours'],df_scores['score_medio'],color='b')
plt.plot(df_scores['n_neighbours'],df_scores['limite_superior'],color='r');
```

```python
df_scores.loc[df_scores.score_medio == df_scores.score_medio.max()]
```

```python
# Utilizamos sklearn para estandarizar la matriz de features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
# Verificamos que las columnas tengan media 0 y desvío 1.
print('media:',np.mean(X_train[:,1]))
print('desvio:',np.std(X_train[:,1]))
```

```python
# Calculamos nuevamente los scores de cross validation,
# pero esta vez sobre los features estandarizados:

scores_para_df_standar = []

for i in range(4,5):
    model = KNeighborsClassifier(n_neighbors=i)
    cv_scores = cross_val_score(model, X_train, y_train, cv=kf)
    dict_row_score = {'score_medio':np.mean(cv_scores),\
                      'score_std':np.std(cv_scores),'n_neighbours':i}
    scores_para_df_standar.append(dict_row_score)
```

```python
df_scores_standar = pd.DataFrame(scores_para_df_standar)
df_scores_standar.head()
```

```python
# Armamos nuevamente los límites para graficar

df_scores_standar['limite_superior'] = df_scores_standar['score_medio'] + df_scores_standar['score_std']
df_scores_standar['limite_inferior'] = df_scores_standar['score_medio'] - df_scores_standar['score_std']
df_scores_standar.head() 
```

```python
df_scores_standar.loc[df_scores_standar.score_medio == df_scores_standar.score_medio.max()]
```

```python
best_k = df_scores_standar.loc[df_scores_standar.score_medio ==\
                               df_scores_standar.score_medio.max(),'n_neighbours'].values
best_k = best_k[0]
best_k
```

```python
# Elegimos el modelo óptimo que nos había indicado cross validation
model = KNeighborsClassifier(n_neighbors=best_k)

# Lo ajustamos sobre datos de entrenamiento
model.fit(X_train,y_train)
```

```python
# Lo utilizamos para predecir en test
X_test = scaler.transform(X_test) # ¡Importantísimo estandarizar también los datos de test!
y_pred = model.predict(X_test)
```

```python
# Obtenemos la matriz de confusión
confusion_matrix(y_test, y_pred)
confusion_knn=confusion_matrix(y_test, y_pred)
```

```python
# Graficamos la matriz de confusión para visualizarla mejor
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='BuGn_r', fmt="d",)
plt.ylabel('Valores verdaderos')
plt.xlabel('Valores predichos');
```

```python
TP_knn = confusion_knn[1, 1]
TN_knn = confusion_knn[0, 0]
FP_knn = confusion_knn[0, 1]
FN_knn = confusion_knn[1, 0]
```

```python
print((TP_knn + TN_knn) / float(TP_knn + TN_knn + FP_knn + FN_knn))
print(accuracy_score(y_test, y_pred))
```

Usando el modelo KNN, para este caso tenemos que este predice correctamente las clases en el orden del 82%

#### Voy a ganar el juicio?
[1 1] Pensaba ganarlo, y lo gane   TT -> 763   
[1 0] Pensaba ganarlo, y lo perdi  FT -> 17  
[0 1] Pensaba no ganarlo, y lo gane  FF -> 1093  
[0 0] Pensaba no ganarlo, y lo perdi TF ->11958

```python
class_error = (FP_knn + FN_knn) / float(TP_knn + TN_knn + FP_knn + FN_knn)

print(class_error)
print(1 - accuracy_score(y_test, y_pred))
```

El modelo tiene un error en la estimacion del 17%, gran parte de este error son los falsos positivos

```python
from sklearn.metrics import recall_score
sensitivity = TP_knn /(float(FN_knn + TP_knn))

print(sensitivity)
print(recall_score(y_test, y_pred))
```

Proporcion de positivos correctamente predichos (5%). Lo que implica que el 95% predijo Falsos Negativos

```python
specificity = TN_knn / (TN_knn + FP_knn)

print(specificity)
```

El 95% de los casos predije que no los iba a ganar y no los gane

```python
from sklearn.metrics import precision_score

precision = TP_knn / float(TP_knn + FP_knn)

print(precision)
print(precision_score(y_test, y_pred))
```

El 13% de los casos que dije que iba a ganar y gané, por ende el 87% erre en la estimacion, creia que los iba a ganar y los perdi

```python
from sklearn.metrics import f1_score

f1 = 2*((precision*sensitivity)/(precision+sensitivity))

print(f1)
print(f1_score(y_test,y_pred))
```

Este modelo contiene errores de precision y de recall. Tienen pesos los falsos negativos como los falsos positivos
